```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

## Functional Method

```python
plt.plot(x, y, '#800080')
plt.xlabel('My X Label')
plt.ylabel('My Y Label')
plt.title('My Graph')
#plt.show()
```

```python
#subplot(nrows, ncols, index)
plt.subplot(1, 2, 1)
plt.plot(x, y, 'g--')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'y*-')
```

```python
plt.subplot(1, 3, 1)
plt.subplot(1, 3, 2)
plt.subplot(1, 3, 3)
```

## Object Oriented Method

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.7, 0.7])
axes.plot(x, y, 'y')
```

```python
canvas = plt.figure()

axes1 = canvas.add_axes([0.1, 0.1, 0.8, 0.8])
axes1.plot(x, y, 'y')

axes2 = canvas.add_axes([0.2, 0.5, 0.3, 0.3])
axes2.plot(x, y, 'g')
```

```python
plt.subplot(2, 3, 1)
plt.subplot(2, 3, 2)
plt.subplot(2, 3, 3)
plt.subplot(2, 3, 4)
plt.subplot(2, 3, 5)
plt.subplot(2, 3, 6)

plt.tight_layout()
```

```python
fig_canvas, fig_axes = plt.subplots(nrows=2, ncols=3)
plt.tight_layout()
```

```python
fig_canvas
```

```python
fig_axes
```

```python
for i in fig_axes:
    for j in i:
        j.plot(x, y, 'g')
    
fig_canvas
```
