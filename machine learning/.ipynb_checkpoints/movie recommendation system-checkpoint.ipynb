```python
import numpy as np
import pandas as pd
import seaborn as sns

%matplotlib inline
```

```python
movies = pd.read_csv('movie_dataset.csv')
```

```python
movies.head()
```

```python
movies.info()
```

```python
movies.describe()
```

```python
features = ['keywords', 'genres', 'cast', 'director']
```

```python
sns.heatmap(movies.isnull(), yticklabels=False)
```

```python
for feature in features:
    movies[feature].fillna(" ", inplace=True)
```

```python
sns.heatmap(movies.isnull(), yticklabels=False)
```

```python
def combineFeatures(row):
    return row['keywords'] + " " + row['genres'] + " " + row['cast'] + " " + \
            row['director']
```

```python
movies['combined_features'] = movies.apply(combineFeatures, axis=1)
```

```python
movies.head()
```

```python
movies['combined_features'].iloc[0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cv = CountVectorizer()
```

```python
countMatrix = cv.fit_transform(movies['combined_features'])
countMatrix
```

```python
cosineSim = cosine_similarity(countMatrix)
cosineSim
```

```python
cosineSim.shape
```

```python
cosineSim[0]
```

```python
def getIndex(title):
    return movies[movies.title == title]['index'].values[0]
```

```python
def getTitle(index):
    return movies[movies.index == index]['title'].values[0]
```

```python
movieUserLikes = input("Enter a movie you like: ")
```

```python
movieIndex = getIndex(movieUserLikes)
movieIndex
```

```python
similarity = list(enumerate(cosineSim[movieIndex]))
similarity
```

```python
sortedMovies = sorted(similarity, key = lambda x : x[1], reverse = True)[1:]
sortedMovies
```

```python
for i in sortedMovies:
    print(getTitle(i[0]))
```
