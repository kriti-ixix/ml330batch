```python
import matplotlib.pyplot as plt
```

```python
classes=['Python','R','Artificial Intelligence','Machine Learning','Data Science']
classes_students=[45,12,78,34,93]
```

```python
plt.figure(figsize=(14,5))
colors=['pink','green','purple','blue','red']
explode=[0,0,0.1,0,0.1]
textprops={'fontsize':15,'color':'m'}
wedgeprops={'linewidth':4,'edgecolor':'black','width':4}

plt.pie(
    classes_students,labels=classes,colors=colors,explode=explode,shadow=True,radius=1.5,
    textprops=textprops,startangle=270,wedgeprops=wedgeprops)
plt.show()
```
