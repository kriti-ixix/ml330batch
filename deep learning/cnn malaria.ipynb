```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
imgGen = ImageDataGenerator(rotation_range=20, width_shift_range=0.1, 
                  height_shift_range=0.1, rescale=1/255,
                  shear_range=0.1, zoom_range=0.1, 
                  horizontal_flip=True, fill_mode='nearest')
```

```python
imageShape = (130, 130, 3)
```

```python
130/5
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPool2D, Dense, Flatten
```

```python
model = Sequential()

model.add(Conv2D(filters=32, kernel_size=(5,5), input_shape=imageShape, activation='relu'))
model.add(MaxPool2D(pool_size=(3,3)))

model.add(Conv2D(filters=32, kernel_size=(5,5), input_shape=imageShape, activation='relu'))
model.add(MaxPool2D(pool_size=(3,3)))

model.add(Conv2D(filters=32, kernel_size=(5,5), input_shape=imageShape, activation='relu'))
model.add(MaxPool2D(pool_size=(3,3)))

model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import EarlyStopping

earlyStopping = EarlyStopping(patience=1)
```

```python
trainImageGen = imgGen.flow_from_directory(myTrainDir, target_size=(130, 130), class_mode='binary')
```

```python
trainImageGen
```

```python
testImageGen = imgGen.flow_from_directory(myTestDir, target_size=(130, 130), 
                                           class_mode='binary', shuffle=False)
```

```python
testImageGen
```

```python
trainImageGen.class_indices
```

```python
model.fit_generator(trainImageGen, validation_data=testImageGen, epochs=20,
                   callbacks=[earlyStopping])
```
