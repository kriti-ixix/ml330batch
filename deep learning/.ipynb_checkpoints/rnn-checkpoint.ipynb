```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 50,  501)
x
```

```python
y = np.sin(x)
y
```

```python
plt.plot(x, y)
```

```python
df = pd.DataFrame(data=y, index=x, columns=['Sine'])
```

```python
df.head()
```

```python
len(df)
```

```python
testPercent = 0.2
```

```python
len(df) * testPercent
```

```python
train = df.iloc[:400]
test = df.iloc[400:]
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(train)
```

```python
train = scaler.transform(train)
test = scaler.transform(test)
```

```python
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
```

```python
generator = TimeseriesGenerator(train, train, length=2, batch_size=1)
```

```python
generator
```

```python
X, y = generator[0]
```

```python
X
```

```python
y
```

```python
generator = TimeseriesGenerator(train, train, length=50, batch_size=1)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, SimpleRNN
```

```python
model = Sequential()

model.add(SimpleRNN(50, input_shape=(50, 1)))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mse')
```

```python
model.summary()
```

```python
model.fit(generator, epochs=5)
```

```python
model.history.history
```

```python
lossDf = pd.DataFrame(model.history.history)
lossDf.head()
```

```python
lossDf.plot()
```
